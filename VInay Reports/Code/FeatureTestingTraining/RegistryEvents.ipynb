```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
import numpy as np
```

```python
a = pd.read_csv("/home/cuckoo/Documents/modified_trainingset.csv")
```

```python
networkdf_train = a.filter(items=['family', 'regkey_read','command_line','regkey_opened','regkey_written'])
networkdf_train
```

```python
c = pd.read_csv("/home/cuckoo/Documents/TestingDatasetModified.csv")
```

```python
networkdf_test = c.filter(items=['Classification', 'regkey_read','command_line','regkey_opened','regkey_written'])
networkdf_test
```

```python
#data cleaning 
x = networkdf_train.drop('family', axis=1)

b = networkdf_train['family']

y = networkdf_test.drop('Classification', axis=1)
d = networkdf_test['Classification']
y = y.drop(y.index[-1])
d = d.drop(d.index[-1])
```

```python
#Random Forest 
rfc = RandomForestClassifier(n_estimators=100, random_state=42)
rfc.fit(x, b)
d_pred = rfc.predict(y)
accuracy = accuracy_score(d, d_pred)
print('Accuracy:', accuracy)
precisionR = precision_score(d, d_pred, pos_label='R')
precisionB = precision_score(d, d_pred, pos_label='B')
print('Precision for Ransomware:', precisionR)
print('Precision for Benign Program:', precisionB)
recallR = recall_score(d, d_pred, pos_label='R')
recallB = recall_score(d, d_pred, pos_label='B')
print('Recall for Ransomware:', recallR)
print('Recall for Benign Program:', recallB)
from sklearn.metrics import f1_score
f1R = f1_score(d, d_pred, pos_label='R')
f1B = f1_score(d, d_pred, pos_label='B')
print('F1-score for Ransomware', f1R)
print('F1-score for Benign Program', f1B)
```
