```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
```

```python
a = pd.read_csv("/home/cuckoo/Documents/modified_trainingset.csv")
a
```

```python
c = pd.read_csv("/home/cuckoo/Documents/TestingDatasetModified.csv")
c
```

```python
#data cleaning 
x = a.drop('family', axis=1)

b = a['family']

y = c.drop('Classification', axis=1)
d = c['Classification']
y = y.drop(y.index[-1])
d = d.drop(d.index[-1])
```

```python
#Random Forest 
rfc = RandomForestClassifier(n_estimators=100, random_state=42)
rfc.fit(x, b)
d_pred = rfc.predict(y)
accuracy = accuracy_score(d, d_pred)
print('Accuracy:', accuracy)
precision = precision_score(d, d_pred, average='micro')
print('Precision:', precision)
recall = recall_score(d, d_pred, average='micro')
print('Recall:', recall)
from sklearn.metrics import f1_score
f1 = f1_score(d, d_pred, average='micro')
print('F1-score', f1)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

#Gradient Boosting Classifier 
gbc = GradientBoostingClassifier(random_state=42)
gbc.fit(x,b)

d_pred = gbc.predict(y)
accuracy = accuracy_score(d, d_pred)
print('Accuracy:', accuracy)

precision = precision_score(d, d_pred, average='micro')
print('Precision:', precision)

recall = recall_score(d, d_pred, average='micro')
print('Recall:', recall)

f1 = f1_score(d, d_pred, average='micro')
print('F1-score', f1)
```

```python
#decision tree

dtc = DecisionTreeClassifier(random_state=42)
dtc.fit(x, b)

d_pred = dtc.predict(y)
accuracy = accuracy_score(d, d_pred)
print('Accuracy:', accuracy)

precision = precision_score(d, d_pred, average='micro')
print('Precision:', precision)

recall = recall_score(d, d_pred, average='micro')
print('Recall:', recall)

f1 = f1_score(d, d_pred, average='micro')
print('F1-score', f1)
```

```python
#svc
from sklearn.svm import SVC

svc = SVC(kernel='rbf', random_state=42)
svc.fit(x, b)

d_pred = svc.predict(y)
accuracy = accuracy_score(d, d_pred)
print('Accuracy:', accuracy)

precision = precision_score(d, d_pred, average='micro')
print('Precision:', precision)

recall = recall_score(d, d_pred, average='micro')
print('Recall:', recall)

f1 = f1_score(d, d_pred, average='micro')
print('F1-score', f1)
```
